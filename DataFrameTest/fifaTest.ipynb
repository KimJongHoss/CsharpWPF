```
private static double solveMean(double[] values)
{
  if (values is null)
    throw new ArgumentNullException(nameof(values));

  if (values.Length == 0)
    throw new ArgumentException($"{nameof(values)} must not be empty");

  double sum = 0;
  for (int i = 0; i < values.Length; i++)
    sum += values[i];

  double mean = sum / values.Length;

  return mean;
}
```

```
foreach (DataFrameColumn col in dataFrame.Columns.Skip(1))
{
    // warning: additional care must be taken for datasets which contain null
    double[] values = Enumerable.Range(0, (int)col.Length).Select(x => Convert.ToDouble(col[x])).ToArray();
    double mean = solveMean(values);
    Console.WriteLine($"{col.Name}   = {mean}(n={values.Length})");
}
```

```
using ScottPlot;
using Microsoft.Data.Analysis;
using System;
using System.Linq;
```

```
var dateArray = dataFrame.Columns["Date"].Cast<DateTime>().Select(date => date.ToOADate()).ToArray();
var ARGArray = dataFrame.Columns["ARG"].Cast<float>().ToArray();
var BRAArray = dataFrame.Columns["BRA"].Cast<float>().ToArray();

ScottPlot.Plot myPlot1 = new();
var argGrape = myPlot1.Add.Scatter(dateArray, ARGArray);
argGrape.LegendText = "ARG";
var braGrape = myPlot1.Add.Scatter(dateArray, BRAArray);
braGrape.LegendText = "BRA";

myPlot1.ShowLegend();
myPlot1.Axes.DateTimeTicksBottom();

myPlot1
```

```
var selectFRAover_2001 = dataFrame.Rows
    .Where(row => ((DateTime)row["Date"]).Year > 2001)
    .Select(row => Convert.ToDouble(row["FRA"]))
    .ToList();

    selectFRAover_2001
```

```
var dateArray2 = dataFrame.Rows
    .Where(row => ((DateTime)row["Date"]).Year > 2001)  // 2001  
    .Select(row => ((DateTime)row["Date"]).ToOADate())  //  double 
    .ToArray();
var FRAArray = selectFRAover_2001.ToArray();
ScottPlot.Plot myPlot2 = new();
myPlot2.Add.Scatter(dateArray2, FRAArray);

ScottPlot.Statistics.LinearRegression reg = new(dateArray2, FRAArray);

Coordinates pt1 = new(dateArray2.First(), reg.GetValue(dateArray2.First()));
Coordinates pt2 = new(dateArray2.Last(), reg.GetValue(dateArray2.Last()));
var line = myPlot2.Add.Line(pt1, pt2);
line.MarkerSize = 0;
line.LineWidth = 2;
line.LinePattern = LinePattern.Solid;

myPlot2.Axes.DateTimeTicksBottom();

myPlot2
```
